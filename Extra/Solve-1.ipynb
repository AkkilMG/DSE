```python
from functools import partial
import matplotlib.pyplot as plt
```

```python
square = lambda x: x*x
derivative = lambda x: 2*x
difference_quotient = lambda f,x,h: (f(x+h)-f(x))/h
derivative_estimate = partial(difference_quotient, square, h=0.00001)
x_value = list(range(-10, 10))
actual_derivative = list(map(derivative, x_value))
estimated_derivatives = list(map(derivative_estimate, x_value))
```

```python
plt.title("Actual Derivatives vs Estimates")
plt.plot(x_value, actual_derivative, 'rx', label="Actual")
plt.plot(x_value, estimated_derivatives, 'b+', label="Estimate")
plt.legend(loc=9)
plt.show()
```

```python
import numpy as np
```

```python
funxtion_to_minimize = lambda x: x**2 + 2*x +1
derivative_of_function = lambda x: 2*x+2
def gradient_descent(initial_x, learning_rate, num_iterations):
    x_value = [initial_x]
    errors = [funxtion_to_minimize(initial_x)]
    for _ in range(num_iterations):
        current_x = x_value[-1]
        new_x = derivative_of_function(current_x)
        x_value.append(new_x)
        errors.append(funxtion_to_minimize(new_x))
    return x_value, errors
def plot_greadient_descent(x_value, errors):
    x=np.linspace(-5, 1, 100)
    y = funxtion_to_minimize(x)
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8,8), sharex=True)
    ax1.plot(x,y, label="Function to Minimize")
    ax1.scatter(x_value, [funxtion_to_minimize(x) for x in x_value], color='red', label='Gradient')
    ax1. set_ylabel('f(x)')
    ax1.legend()
    ax1.set_title('Gradient Descent Illustration')
    ax2.plot(errors, marker='o', color="green")
    ax2. set_xlabel('Iteration')
    ax2. set_ylabel('Error')
    ax2.set_title('Error Reduction Over Iterations')
    plt.tight_layout()
    plt.show()
initial_x, learning_rate, num_iterations = -4.0, 0.54, 10
path, errors = gradient_descent(initial_x, learning_rate, num_iterations)
plot_greadient_descent(path, errors)
```
